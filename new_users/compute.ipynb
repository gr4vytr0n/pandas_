```python
import pandas as  pd
import numpy as np
%matplotlib inline
```

```python
# get the sum and length of a group
df = pd.DataFrame({
    'group1': ['a', 'a', 'b', 'b'],
    'value': np.arange(10, 50, 10)
})
df
```

```python
group = df.groupby('group1')
group.agg([len, sum])
```

```python
# add a column that is equal to the sum of a group
df = pd.DataFrame({
    'labels': ['a', 'a', 'b', 'b'],
    'value': np.arange(10, 50, 10)
})
df
```

```python
group = df.groupby('labels')['value']
df['value.sum'] = group.transform('sum')
df
```

```python
# get the month name out of a data column
df = pd.DataFrame({
    'col1': [
        pd.Timestamp('20130102000030'),
        pd.Timestamp('2013-01-03 00:00:30'),
        pd.Timestamp('1/4/2013 000030')
    ]
})
df
```

```python
df['Month Number'] = df['col1'].apply(lambda x: x.month)
df['Day'] = df['col1'].apply(lambda x: x.day)
df['Month Name'] = df['col1'].apply(lambda x: x.strftime('%B'))
df['Week Day'] = df['col1'].apply(lambda x: x.strftime('%A'))
df
```

```python
# create a column based on two other columns
df = pd.DataFrame({
    'col1': ['minus', 'minus', 'positive', 'nan'],
    'col2': np.arange(10, 50, 10)
})
df
```

```python
df['col3'] = df['col2'] * df['col1'].apply(lambda x: -1 if x == 'minus' else (1 if x == 'positive' else np.nan))
df
```

```python
# apply function to a group and add results to original data frame
df = pd.DataFrame({
    'group1': ['a', 'a', 'a', 'b', 'b', 'b'],
    'group2': ['c', 'c', 'd', 'd', 'd', 'e'],
    'value1': [1.1, 2, 3, 4, 5, 6],
    'value2': [7.1, 8, 9, 10, 11, 12]
})
df
```
