```python
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('../train.csv')
df.columns
```

```python
# plot of counts of a specific column of data
df.Survived.value_counts().plot(kind='barh')
```

```python
# bar plot median values
df.groupby('Survived')['Pclass'].agg(np.median).plot(kind='bar')
```

```python
# box plot
g = sns.factorplot("Survived", "Fare", "Pclass", df, kind="box", palette="PRGn", aspect=2.25)
g.set(ylim=(df['Fare'].min(), df['Fare'].max() + 100))
```

```python
sns.factorplot("SibSp", "Pclass", data=df, hue="Survived", size=3, aspect=2)
```

```python
# violin plot
sns.violinplot(df["Survived"], df["Pclass"], color="Red").set_ylim(-1, df['Pclass'].max()+3)
sns.despine(left=True)
```

```python
# regression plot
sns.regplot("Survived", "Fare", data=df, robust=True, ci=95, color="seagreen")
sns.despine()
```

```python
# bar plot of median values
sns.factorplot("Survived", "Pclass", data=df, palette="PuBu_d", estimator=np.median)
```
