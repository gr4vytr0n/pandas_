```python
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_csv('../train.csv')
```

```python
# melt data
melt = pd.melt(df, id_vars=['PassengerId', 'Survived', 'Pclass'])
melt.iloc[0:5, :]
```

```python
# descriptive statistics of dataset
df.describe()
```

```python
# crosstab of data by specified columns
pd.crosstab(df['Pclass'], df['Survived'])
```

```python
# return subset of data
pclass_subset = df[df['Pclass'] == 1]
pclass_subset.iloc[0:3]
```

```python
# group data and obtain mean
df.groupby(['Survived', 'Pclass']).mean()
```

```python
# check boolean condition
(df['Fare'] > 50).any()
```

```python
# get descriptive statistics about specific column
df.Fare.describe()
```

```python
# group data and obtain the mean values
df.groupby(['Survived', 'Pclass']).aggregate(np.mean)
```

```python
# group data and get sums
df.groupby(['Survived', 'Pclass']).aggregate(np.sum)
```

```python
# return boolean values for a specified criteria
df['Survived'].map(lambda x: x == 1).iloc[0:10]
```
