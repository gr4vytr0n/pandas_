```python
# joining
```

```python
# merge method allows joining two DataFrames on one or more keys
# parameters on, left_on, right_on, left_index, right_index specify columns or indices to join on
# how parameter can change merge methods default of inner join
left_frame = pd.DataFrame({'key': range(5), 'left_value': ['a', 'b', 'c', 'd', 'e']})
left_frame
```

```python
right_frame = pd.DataFrame({'key': range(2, 7), 'right_value': ['f', 'g', 'h', 'i', 'j']})
right_frame
```

```python
# inner join
pd.merge(left_frame, right_frame, on='key', how='inner')
```

```python
pd.merge(left_frame, right_frame, on='key', how='outer')
```

```python
# combining
```

```python
# pandas concat method combines DataFrames along an given axis
```

```python
pd.concat([left_frame, right_frame])
```

```python
pd.concat([left_frame, right_frame], axis=1)
```

```python
# use pandas groupby method
```

```python
headers = ['name', 'title', 'department', 'salary']
salaries = pd.read_csv('city-of-chicago-salaries.csv', header=0, names=headers,
                       converters={'salary': lambda x: float(x.replace('$', ''))})
salaries.head()
```

```python
# pandas groupby method returns special object with many useful methods
```

```python
by_dept = salaries.groupby('department')
by_dept
```

```python
# count returns total number of not null values in the column
by_dept.count().head()
```

```python
# size returns total number of records in each group
by_dept.size().head()
```

```python
# summation
by_dept.sum().head()
```

```python
# mean
by_dept.mean().head()
```

```python
# median
by_dept.median().head()
```

```python
# operations can be done on individual Series within a grouped object
```

```python
# five departments with the most distinct titles
by_dept.title.nunique().sort_values(ascending=False)[:4]
```

```python
# split-apply-combine
```

```python
# highest paid employee within each department
```

```python
def ranker(df):
    ''' apply rank to each employee based upon their salary '''
    df['dept_rank'] = np.arange(len(df)) + 1
    return df
```

```python
# use apply method to apply ranker function to each group
salaries.sort_values('salary', ascending=False, inplace=True)
salaries = salaries.groupby('department').apply(ranker)
salaries[salaries.dept_rank == 1].head(5)
```
